```python

# Get average word length note we are not removing words
def GetWordLength(text):
    tokens = nltk.word_tokenize(text.lower())
    averageLen = []
    averageLen= [len(x) for x in tokens]
    return np.sum(averageLen)/len(tokens)

def getPOSCounts(text, desiredPOS = ['NN','NNS','NNP','NNPS','VB','VBG','VBD','VBN','VBP','VBZ','JJ','JJR','JJS']):
    text = text.split()
    tokens_tag = pos_tag(text)
    newList = [x[1] for x in tokens_tag if x[1] in desiredPOS]
    c = Counter(newList)

    res = Counter({k: v for k, v in c.items() if v > 1})

    array = []
    for tag in desiredPOS:
        if tag in res:
            array.append(res[tag])
        else:
            array.append(0)
    return array




desiredPOS = ['NN','NNS','NNP','NNPS','VB','VBG','VBD','VBN','VBP','VBZ','JJ','JJR','JJS']

columnsToKeep = ['reviewId', 'userName', 'content', 'score',
       'thumbsUpCount', 'at', 'title', 'description', 'summary', 'size',
       'developer', 'contentRating','updated','released', 'appId', 'sentiment']

mypath = 'sentReviews/'

from os import listdir, path
from os.path import isfile, join
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

reviews_df = None
for app_id in onlyfiles:

    temp_df = None
    if path.exists(mypath + app_id):
        temp_df = pd.read_pickle(mypath + app_id, compression='gzip')
        
    else:
        continue
        
#     display(temp_df.columns)
    temp_df=temp_df[columnsToKeep]
    
    
    # now let's modify this df with the word counts...
    temp_df['review_word_count'] = temp_df['content'].apply(lambda x: len(nltk.word_tokenize(x.lower())))
    temp_df['review_average_word_len'] = temp_df['content'].apply(lambda x: GetWordLength(x))


    
    pos_ds = temp_df['content'].apply(lambda x: getPOSCounts(x,desiredPOS))
    split_df = pd.DataFrame(pos_ds.to_list(), columns = desiredPOS)

    temp_df = pd.concat([temp_df, split_df], axis=1)
    
    temp_df.to_pickle(str(mypath + app_id),  
                 compression={'method': 'gzip', 'compresslevel': 9}, 
                 protocol=4
                )
```
