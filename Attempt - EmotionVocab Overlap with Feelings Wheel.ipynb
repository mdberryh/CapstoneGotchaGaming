```python
emo_ind = []
emo_core_type = []
emo_sub_type = []
```

```python
for v in emo_vocab:
  vstem = sb_stemmer.stem(v)
  if vstem in wheel_stem:
    emo_ind.append(wheel[wheel['stem'] == vstem].index.tolist()[0])
    emo_sub_type.append(wheel[wheel['stem'] == vstem]['Sub Type'].tolist()[0])
    emo_core_type.append(wheel[wheel['stem'] == vstem]['Core Type'].tolist()[0])
  else:
    emo_ind.append('-')
    emo_core_type.append('-')
    emo_sub_type.append('-')
```
